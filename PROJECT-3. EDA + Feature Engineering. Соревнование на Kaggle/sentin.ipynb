```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import time
nltk.downloader.download('vader_lexicon')

sent_analyzer = SentimentIntensityAnalyzer()

rev1 = "The hotel was very good, I love it!"
rev2 = "It was just horrible, the worst ever."

print(f"review 1:\n{rev1}\nScore: {sent_analyzer.polarity_scores(rev1)}")

print(f"\nreview 2:\n{rev2}\nScore: {sent_analyzer.polarity_scores(rev2)}")
```

```python
import pandas as pd
hotels = pd.read_csv('data/hotels.csv')
hotels.head(2)
```

```python
review = hotels[['negative_review','positive_review']]
review['negative_review'].replace('No Negative','Nothing',inplace=True)
review['positive_review'].replace('No Positive','Nothing',inplace=True)
#review = review.replace(['No Negative', 'Nice'],['No Positive', 'Bad'])
review.head(30)
```

```python
def extract_value(x):
    out = sent_analyzer.polarity_scores(x)
    return out['neg'],out['neu'],out['pos']
```

```python
review['negative_review_set'] = review['negative_review'].apply(lambda x: extract_value(x)[0])
review['positive_review_set'] = review['positive_review'].apply(lambda x: extract_value(x)[2])
```
